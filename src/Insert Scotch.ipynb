```python
%matplotlib inline
import pprint
from pprint import pprint
import re
import os
from os import listdir
from os.path import isfile, join, walk
import csv
from storm.locals import *
from collections import Counter
from matplotlib import pyplot as plt
import math
import statistics
import numpy as np
import pandas as pd
import seaborn as sns

class RETAILER(object):
    __storm_table__ = "RETAILER"
    RETAILER_ID = Int(primary=True)
    NAME = Unicode()
    TYPE = Unicode()
    CITY = Unicode()
    STATE = Unicode()
    COUNTRY = Unicode()
    SHIPPING_COST = Float()
    ADDITIONAL_COST = Float()
    
      
    def __init__(name, retailer_type, city, state, country, shipping_cost, additional_cost):
        # self.RETAILER_ID = int(retailer_id)
        self.NAME = u'' + name
        self.TYPE = u'' + retailer_type
        self.CITY = u'' + city
        self.STATE = u'' + state
        self.COUNTRY = u'' + country
        self.SHIPPING_COST = float(shipping_cost)
        self.ADDITIONAL_COST = float(additional_cost)

    def __str__(self):
        return 'Player[ RETAILER_ID = ' + str(self.RETAILER_ID) + ', NAME = ' + self.NAME + ', TYPE = ' + self.TYPE + ', CITY = ' + self.CITY + ', STATE = ' + self.STATE + ', COUNTRY = ' + self.COUNTRY + ', SHIPPING_COST = ' + str(self.SHIPPING_COST) + ', ADDITIONAL_COST = ' +  str(self.ADDITIONAL_COST) + ']'

    
class DISTILLERY(object):
    __storm_table__ = "DISTILLERY"
    DISTILLERY_ID = Int(primary=True)
    NAME = Unicode()
    REGION = Unicode()
      
    def __init__(name, region):
        # self.DISTILLERY_ID = int(distillery_id)
        self.NAME = u'' + name
        self.REGION = u'' + region

    def __str__(self):
        return 'Player[ DISTILLERY_ID = ' + str(self.DISTILLERY_ID) + ', NAME = ' + self.NAME + ', REGION = ' + self.REGION + ']'

    
class PRODUCER(object):
    __storm_table__ = "PRODUCER"
    PRODUCER_ID = Int(primary=True)
    NAME = Unicode()
      
    def __init__(name):
        # self.PRODUCER_ID = int(producer_id)
        self.NAME = u'' + name

    def __str__(self):
        return 'Player[ PRODUCER_ID = ' + str(self.PRODUCER_ID) + ', NAME = ' + self.NAME + ']'

    
class SCOTCH(object):
    __storm_table__ = "SCOTCH"
    SCOTCH_ID = Int(primary=True)
    PRODUCER = Int()
    DISTILLERY = Int()
    NAME = Unicode()
    AGE = Int()
    
    def __init__(producer, distillery, name, age):
        # self.SCOTCH_ID = int(scotch_id)
        self.PRODUCER = int(producer)
        self.DISTILLERY = int(distillery)
        self.NAME = u'' + name
        self.AGE = int(age)

    def __str__(self):
        return 'Player[ SCOTCH_ID = ' + str(self.SCOTCH_ID) + ', PRODUCER = ' + self.PRODUCER + ', DISTILLERY = ' + self.DISTILLERY + ', NAME = ' + self.NAME + ', AGE = ' + self.AGE + ']'

class SCOTCH(object):
    __storm_table__ = "PRICE"
    PRICE_ID = Int(primary=True)
    RETAILER = Int()
    SCOTCH = Int()
    PRICE = Float()
    
    def __init__(retailer, scotch, price):
        # self.SCOTCH_ID = int(scotch_id)
        self.RETAILER = int(retailer)
        self.SCOTCH = int(scotch)
        self.PRICE = float(price)

    def __str__(self):
        return 'Player[ PRICE_ID = ' + str(self.PRICE_ID) + ', RETAILER = ' + self.RETAILER + ', SCOTCH = ' + self.SCOTCH + ', PRICE = ' + self.PRICE + ']'
    

MATH_8756_HOME = os.environ.get('MATH_8756_HOME')
CSV_DIR = MATH_8756_HOME + '/' + 'clean data'
DB_DIR = MATH_8756_HOME + '/' + 'db'
DB = DB_DIR + '/scotch_prices.db'

# This is what we're trying to populate

print 'opening ' + DB
scotch_db = create_database('sqlite:' + DB)
scotch_store = Store(scotch_db)

for root, subdirs, files in os.walk(CSV_DIR):
    for fileName in files:
        if (fileName == 'retailers.csv'):
            continue
        with open(CSV_DIR + '/' + fileName, 'rb') as csvfile:
            retailer = 0
            if (fileName == 'ace_spirits.csv'):
                retailer = 1
            elif (fileName == 'astor_wines.csv'):
                retailer = 2
            elif (fileName == 'beltramos.csv'):
                retailer = 3
            elif (fileName == 'binnys.csv'):
                retailer = 4
            elif (fileName == 'caskers.csv'):
                retailer = 5
            elif (fileName == 'd_and_m.csv'):
                retailer = 6
            elif (fileName == 'drink_up_ny.csv'):
                retailer = 7
            elif (fileName == 'hi_time_wine.csv'):
                retailer = 8
            elif (fileName == 'klwines.csv'):
                retailer = 9
            elif (fileName == 'liquorama.csv'):
                retailer = 10
            elif (fileName == 'love_scotch.csv'):
                retailer = 11
            elif (fileName == 'pacific_online_spirits.csv'):
                retailer = 12
            elif (fileName == 'quality_liquor_store.csv'):
                retailer = 13
            elif (fileName == 'sfwtc.csv'):
                retailer = 14
            elif (fileName == 'the_whisky_exchange.csv'):
                retailer = 16
            elif (fileName == 'ultimate_wine_shop.csv'):
                retailer = 17
            elif (fileName == 'whisky_shop_usa.csv'):
                retailer = 18
            elif (fileName == 'wine_and_cheese_place.csv'):
                retailer = 19
                
            whisky = csv.reader(csvfile, delimiter=',')
            for row in whisky:
                print str(retailer)
                print fileName
                print row
                



    
scotch_store.flush()
scotch_store.commit()

        

#target.write(str(flames_lames_dict))
#target.close()
```
